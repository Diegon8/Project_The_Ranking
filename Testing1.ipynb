```python
import os
from dotenv import load_dotenv
import json
import requests
load_dotenv()
```

```python

def requestGitHub (endpoint, token=os.getenv("apikey"),queryParams={}):
    
    baseUrl='https://api.github.com'
    
    url = baseUrl+endpoint

    headers = {
        "Authorization": f"token {token}"}

    res = requests.get(url, headers=headers)
    
    data = res.json()
    
    if res.status_code != 200:
        raise ValueError(f'Invalid github api call: {data["message"]}')
    
    return data
```

```python
url_pulls = '/repos/ironhack-datalabs/datamad0820/pulls'

requestGitHub(url_pulls)


```
